```python
 # Para colunas de texto (categóricas), pode-se preencher com a moda (valor mais frequente) ou uma constante como 'Desconhecido'.
 # Definimos que substiutiriamos por Desconhecido

if 'descricao' in df.columns and df['descricao'].isnull().any():
  moda_descricao = "Desconhecido"
df['descricao'].fillna(moda_descricao, inplace=True)
print(f"Valores ausentes em 'descrição' preenchidos com a moda: {moda_descricao}")
```

```python
# Estratégia B: Preencher valores ausentes do iptu com um valor específico (0)

if 'iptu' in df.columns and df['iptu'].isnull().any():
 moda_iptu = 0.0
 df['iptu'].fillna(moda_iptu, inplace=True)
 print(f"\nValores ausentes em 'iptu' preenchidos com a mediana: {moda_iptu}")
```

```python
# Estratégia B: Preencher valores ausentes do "Valor" com um valor específico (0)

if 'valor' in df.columns and df['valor'].isnull().any():
 moda_valor = 0.0
 df['valor'].fillna(moda_valor, inplace=True)
 print(f"\nValores ausentes em 'valor' preenchidos com a mediana: {moda_valor}")
```

```python
# Estratégia B: Preencher valores ausentes do "Taxa_condominial" com um valor específico (0)

if 'taxa_condominial' in df.columns and df['taxa_condominial'].isnull().any():
 moda_taxa_condominial = 0.0
 df['taxa_condominial'].fillna(moda_taxa_condominial, inplace=True)
 print(f"\nValores ausentes em 'taxa_condominial' preenchidos com valor 0: {moda_taxa_condominial}")
```

```python
import pandas as pd

caminho_do_arquivo = '/content/tb_mercadoimob.csv'
df = pd.read_csv(caminho_do_arquivo)
```

```python
# --- TRATANDO VALORES AUSENTES (NaN) --
# 1. Contar quantos valores ausentes existem em cada coluna
print("--- Contagem de valores ausentes por coluna: ---")
print(df.isnull().sum())

if 'descricao' in df.columns and df['descricao'].isnull().any():
  moda_descricao = "Desconhecido"
  df['descricao'] = df['descricao'].fillna(moda_descricao) # Use assignment instead of inplace=True

if 'iptu' in df.columns and df['iptu'].isnull().any():
  moda_iptu = 0.0
  df['iptu'] = df['iptu'].fillna(moda_iptu) # Use assignment instead of inplace=True

if 'valor' in df.columns and df['valor'].isnull().any():
  moda_valor = 0.0
  df['valor'] = df['valor'].fillna(moda_valor) # Use assignment instead of inplace=True

if 'taxa_condominial' in df.columns and df['taxa_condominial'].isnull().any():
  moda_taxa_condominial = 0.0
  df['taxa_condominial'] = df['taxa_condominial'].fillna(moda_taxa_condominial) # Use assignment instead of inplace=True

# Remover as colunas que não terão impacto na nossa analise: 'contato' do DataFrame
columns_to_drop = ['contato', 'estado_construcao', 'date_part', 'id_fonte', 'link_anuncio', 'fonte', 'tipo_uso', 'imovel_lancamento', 'bl_temporada', 'data_raspagem', 'hash', 'bl_ativo', 'estado_sigla', 'cidade', 'bairro_tratado', 'complemento', 'precisao', 'ponto', 'data_tabela', 'is_outlier', 'logradouro', 'numero', 'lat', 'lon', 'latlong']
df.drop(columns=columns_to_drop, inplace=True)

# Remover linhas que contêm qualquer valor ausente no 'cep'
df.dropna(subset=['cep'], inplace=True)

#Vamos deletar os dados correspondentes a aluguel na coluna tipo_negocio
df = df[df['tipo_negocio'] != 'Aluguel']

print("\n--- Contagem de valores ausentes por coluna após tratamento: ---")
print(df.isnull().sum())
```
